### CH1 : YOLO basics

```python
import torch 
import scipy
import ultralytics
import cv2
import math
import cvzone
```

```python
from ultralytics import YOLO
```

```python

model = YOLO("Yolo-Weights/yolov8n.pt")
```

```python
results =model.predict("D:/ASL/PROJETs/ML/Object_Detetction/1. Running_YOLO/images/Nbikes.jpg" ,stream =True)
for r in results :
    all_classes = list(r.names.values())
    #print(f"r.boxes => {r}\n")
    for box in r.boxes:
        print(box)
        #print(f"Number Of Objects in the Image :\n{len(r.boxes)} ")
        #print(f"\nFirst Box Cordiantes : \n{box.xyxy[0]} ")
        class_id =box.cls
        print(f"the Clasee >> {class_id[0]} ")
        break
        
#frame =results[0].plot()
#cv2.imshow("frame",frame)
#cv2.waitKey(0)  
```

##### Collecte the classes
* We're Gonna write alll the classes in a file.txt so we can use the class names Later on  

```python
print(all_classes)
with open("80Classes.txt","w") as f :
    for cls in all_classes :
        f.write(cls + "\n")
```

### CH2 : YOLO With Webcam 
* TEST the functionalities using the webcam 

```python
cap = cv2.VideoCapture("./videos/cars.mp4") # Open The Video = '0'
cap.set(cv2.CAP_PROP_FRAME_WIDTH,1280) # Width is equal to 1280 pixels 
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,720)
while True:
    success,frame = cap.read()
    results =model.predict(frame, stream=True)
    for r in results :
          boxes = r.boxes # Len Boxes = Number Of Objects
          for box in boxes:
              print(box)
              x1,y1,x2,y2 = box.xyxy[0]
              x1,y1,x2,y2 = int(x1),int(y1),int(x2),int(y2)
              w, h = x2-x1, y2-y1
              #
              # the Class Name 
              cls = int(box.cls[0])
              class_detected = all_classes[cls]
              #The Confidence Sore 
              conf = math.ceil(box.conf[0]*100)/100 
              # Display only the Permitted classes
              if class_detected in ["bicycle","car"] and conf >0.5 :    
                cvzone.putTextRect(frame,f"{class_detected} {conf} ",(x1,y1+30),colorR=(0,255,0),colorB=(0,0,0),scale=0.8,thickness=1,offset=5)
                #print(f"First Box Cordiantes : \n{x1} {y1} {x2} {y2} ")
                cvzone.cornerRect(frame,(x1,y1,w,h),colorR=(255,255,255),rt=1,l=10)
    cv2.imshow("frame",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
            cv2.destroyAllWindows()
            cap.release()
            break
        
```
